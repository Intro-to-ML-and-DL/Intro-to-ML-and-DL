```python
from sklearn.datasets import make_blobs
import numpy as np
import matplotlib.pyplot as plt
from numpy import where
```

```python
X_test, y_test = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=1, cluster_std=4)
```

```python
for j in range(2):
 i = where(y_test == j)
 plt.scatter(X_test[i, 0], X_test[i, 1])
plt.show()
```
