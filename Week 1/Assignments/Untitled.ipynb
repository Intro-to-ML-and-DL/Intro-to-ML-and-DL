```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Set random seed for reproducibility
np.random.seed(0)

# Number of samples and features
num_samples = 1000
num_features = 5

# Generate random feature data
X = np.random.randn(num_samples, num_features)

# Generate random target data
true_weights = np.random.randn(num_features)
true_bias = np.random.randn()
y = np.dot(X, true_weights) + true_bias

# Add some noise to the target data
y += np.random.randn(num_samples)

# Visualize the data
fig, axs = plt.subplots(1, num_features, figsize=(20, 4))

for i in range(num_features):
    axs[i].scatter(X[:, i], y, s=5, color='blue', label='Data')
    axs[i].set_xlabel(f'Feature {i+1}')
    axs[i].set_ylabel('Target')
    axs[i].legend()

    # Fit linear regression model
    model = LinearRegression()
    model.fit(X[:, i].reshape(-1, 1), y)

    # Predict target values using the fitted model
    predictions = model.predict(X[:, i].reshape(-1, 1))

    # Plot the best-fit line
    axs[i].plot(X[:, i], predictions, color='red', label='Best Fit Line')
    axs[i].legend()

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make the data
np.random.seed(3)
x = 4 + np.random.normal(0, 2, 24)
y = 4 + np.random.normal(0, 2, len(x))
# size and color:
sizes = np.random.uniform(15, 80, len(x))
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors, vmin=0, vmax=100)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

x, y = make_regression(n_samples=1000, n_features=5)

colors = np.random.uniform(10, 900, len(x))
fig, axs = plt.subplots(1, 5, figsize=(20, 4))

reg = LinearRegression().fit(x, y)

for i in range(5):
    axs[i].scatter(x[:, i], y, c=colors)
    axs[i].set_xlabel(f"Feature {i+1}")
    axs[i].set_ylabel("Target")

    # Predict target values for all features
    probable_target = reg.predict(x)

    axs[i].plot(x[:, i], probable_target, color='red', label='Best Fit Line')

plt.show()
```
