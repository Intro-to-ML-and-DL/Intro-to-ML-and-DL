```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X_test, y_test = make_regression(n_samples=1000, n_features=5, noise=0)
```

```python
rows, cols = (1000, 1)
X1_test = [[0 for i in range(cols)] for j in range(rows)]
for i in range(1000):
  X1_test[i][0]=X_test[i][0]
l1 = LinearRegression()
l1.fit(X1_test, y_test)
Y1_pred = l1.predict(X1_test)
plt.scatter(X1_test,y_test)
plt.plot(X1_test, Y1_pred, color='red')
plt.show()
```
