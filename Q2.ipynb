```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
from numpy import where
```

```python
X_test, y_test = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, n_classes=2, n_clusters_per_class=2)
```

```python
for j in range(2):
 i = where(y_test == j)
 plt.scatter(X_test[i, 0], X_test[i, 1])
plt.show()
```
