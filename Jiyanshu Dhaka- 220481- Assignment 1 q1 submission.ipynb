```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression as mr
from sklearn.linear_model import LinearRegression

x, y = mr(n_samples=1000, n_features=5)

fig, axs = plt.subplots(1, 5, figsize=(15, 3))  
for i in range(5):
    axs[i].scatter(x[:, i], y, s=5)
    axs[i].set_xlabel(f"Feature {i+1}")
    axs[i].set_ylabel("Target")

    reg = LinearRegression().fit(x[:, i].reshape(-1, 1), y)
    probable_target = reg.predict(x[:, i].reshape(-1, 1))
    axs[i].plot(x[:, i], probable_target, color='green', label='Best Fit Line')
    axs[i].legend()

plt.tight_layout()
plt.show()
```
